# Correlation Analysis:

Use heatmaps to examine correlations between variables.

For Togo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Timestamp' column to datetime
df_togo['Timestamp'] = pd.to_datetime(df_togo['Timestamp'])

# Calculate the correlation matrix for the Togo dataset
correlation_matrix_togo = df_togo.corr()

# Set the figure size for the heatmap
plt.figure(figsize=(14, 10))

# Plot the heatmap with improved aesthetics
sns.heatmap(correlation_matrix_togo, annot=True, fmt=".2f", cmap="coolwarm", 
            annot_kws={"size": 10}, linewidths=0.5, linecolor='gray')

# Set the title with increased font size
plt.title('Correlation Matrix (Togo)', fontsize=16)

# Display the heatmap
plt.show()
```

# Wind Analysis:

Use polar plots to analyze wind speed and direction

For Togo

```python
print(df_togo.columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt

df_togo.rename(columns={
    'WS': 'Wind_Speed',
    'WD': 'Wind_Direction',  # Rename 'WD' to 'Wind_Direction'
    'WSgust': 'Max_Wind_Gust_Speed',
    'WSstdev': 'Wind_Speed_Stdev',
    'Wind_Direction_Stdev': 'Wind_Direction_Stdev'
}, inplace=True)


if 'Wind_Direction' in df_togo.columns:
    theta = np.radians(df_togo['Wind_Direction'])
else:
    print("Column 'Wind_Direction' not found in DataFrame")


# Get wind speed values
r = df_togo['Wind_Speed']

# Create a polar plot
plt.figure(figsize=(8, 8))
plt.polar(theta, r, 'o', markersize=3, alpha=0.75)
plt.title('Wind Speed and Direction (Togo)')
plt.show()
```

# Temperature Analysis:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Rename columns if needed
df_togo.rename(columns={
    'WS': 'Wind_Speed',
    'WD': 'Wind_Direction',
    'WSgust': 'Max_Wind_Gust_Speed',
    'WSstdev': 'Wind_Speed_Stdev',
    'Wind_Direction_Stdev': 'Wind_Direction_Stdev'
}, inplace=True)

# Temperature Analysis: Examine how RH influences temperature and solar radiation
plt.figure(figsize=(10, 5))
plt.scatter(df_togo['RH'], df_togo['Tamb'], alpha=0.5)
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Ambient Temperature (C)')
plt.title('RH vs Ambient Temperature (Togo)')
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(df_togo['RH'], df_togo['GHI'], alpha=0.5)
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Global Horizontal Irradiance (W/m)')
plt.title('RH vs GHI (Togo)')
plt.show()
```

# Z-Score Analysis

For Togo

```python
# Calculate Z-scores
z_scores = stats.zscore(df_togo[['GHI', 'DNI', 'DHI']])
df_z_scores = pd.DataFrame(z_scores, columns=['GHI', 'DNI', 'DHI'])

# Display the first few rows of the Z-scores DataFrame
print("Z-Scores:")
print(df_z_scores.head())
print("Summary Statistics of Z-Scores:")
print(df_z_scores.describe())
# Histogram of Z-Scores
df_z_scores.hist(bins=30, figsize=(10, 6))
plt.suptitle('Histogram of Z-Scores')
plt.show()
```
